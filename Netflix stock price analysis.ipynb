```python
df.drop('Adj Close',axis=1, inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
plt.figure(figsize=(15, 10))
plt.plot(df['Close'])
```

```python
df['Close'].shape
```

```python
df1 = df[['Close']]
df1.head()
```

```python
# Create a variable to predict a certain number of days into the future ('x').
future_days = 252
```

```python
# Create a column that will store target data that will be used to predict future values of the stock closing price.
df1['prediction'] = df1[['Close']].shift(-future_days)
```

```python
df1['prediction'].tail()
```

```python
df1.columns
```

```python
# Create feature data 'X' and remove the future days
X = np.array(df1.drop('prediction',axis=1)[:-future_days])
```

```python
X
```

```python
# create target data 'y' and get all target values except the future days
y = np.array(df1['prediction'])[:-future_days]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
Forest = RandomForestRegressor().fit(X_train, y_train)
lm = LinearRegression().fit(X_train, y_train)
K = KNeighborsRegressor(n_neighbors=5).fit(X_train, y_train)
```

```python
# get the last 'x' rows to the feature dataset
x_future = df1.drop('prediction', axis=1)[-future_days:]
x_future = x_future.tail(future_days)
x_future = np.array(x_future)
```

```python
x_future
```

```python
Forest_prediction = Forest.predict(x_future)
lm_prediction = lm.predict(x_future)
K_prediction = K.predict(x_future)
```

```python
valid = df1[X.shape[0]:]
valid['prediction'] = Forest_prediction

plt.figure(figsize=(20, 12))
plt.title('Netflix_data')
plt.xlabel('days')
plt.ylabel('Close prices in USD')
plt.plot(df['Close'])
plt.plot(valid[['Close',  'prediction']])
plt.legend(['original_data', 'valid_data', 'predicted_data'])
```

```python
valid = df1[X.shape[0]:]
valid['prediction'] = lm_prediction

plt.figure(figsize=(20, 12))
plt.title('Netflix_data')
plt.xlabel('days')
plt.ylabel('Close prices in USD')
plt.plot(df['Close'])
plt.plot(valid[['Close',  'prediction']])
plt.legend(['original_data', 'valid_data', 'predicted_data'])
```

```python
valid = df1[X.shape[0]:]
valid['prediction'] = K_prediction

plt.figure(figsize=(20, 12))
plt.title('Netflix_data')
plt.xlabel('days')
plt.ylabel('Close prices in USD')
plt.plot(df['Close'])
plt.plot(valid[['Close',  'prediction']])
plt.legend(['original_data', 'valid_data', 'predicted_data'])
```

Among the models used, based on the graphs shown above, the linear regression model gives the best results
